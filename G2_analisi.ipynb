### TESTO UTILIE
Poiché ogni partita è identificata univocamente da gameid ma compare più volte nel dataset per rappresentare statistiche di team e giocatori, il numero di partite è stato calcolato considerando i valori unici di gameid.
Successivamente, ogni partita è stata assegnata al relativo split (Spring, Summer, Finals) e aggregata per ottenere la distribuzione delle partite nel corso della stagione 2025.


```python
summary_gamesPerSplit_df = games_per_split.reset_index()
summary_gamesPerSplit_df.columns = ['split', 'games']
total_games = summary_gamesPerSplit_df['games'].sum()

summary_gamesPerSplit_df['percentage (%)'] = (
    summary_gamesPerSplit_df['games'] / total_games * 100
).round(2)

summary_gamesPerSplit_df.head()
```

```python

labels = summary_gamesPerSplit_df['split']
sizes = summary_gamesPerSplit_df['games']

fig, ax = plt.subplots(figsize=(7, 7))

ax.pie(
    sizes,
    labels=labels,
    autopct='%1.1f%%',
    startangle=90,
    wedgeprops={'width': 0.4},
    pctdistance=0.80
)

# Numero totale al centro
ax.text(
    0, 0,
    f'{total_games}\nTotal Games',
    ha='center',
    va='center',
    fontsize=12,
    fontweight='bold'
)

ax.set_title(
    'Distribuzione delle partite LEC 2025 per Split',
    fontsize=14
)

plt.show()
```

**Qual è stata la durata media delle partite LEC nel 2025?**

```python
games_duration_df = lec_df.drop_duplicates(subset='gameid').copy()

games_duration_df['gamelength_min'] = games_duration_df['gamelength'] / 60
mean_duration = games_duration_df['gamelength_min'].mean()
median_duration = games_duration_df['gamelength_min'].median()

n_total = games_duration_df['gamelength_min'].shape[0]
n_over_45 = (games_duration_df['gamelength_min'] > 45).sum()
pct_over_45 = n_over_45 / n_total * 100
```

```python
plt.figure(figsize=(8, 5))
ax = sns.histplot(games_duration_df['gamelength_min'], bins=20, kde=True)

# Linee verticali: media e mediana
ax.axvline(mean_duration, linestyle='--', linewidth=2, label=f"Media: {mean_duration:.1f} min")
ax.axvline(median_duration, linestyle=':', linewidth=2, label=f"Mediana: {median_duration:.1f} min")

# Asse Y con step fisso (come hai già fatto)
y_max = ax.get_ylim()[1]
ax.set_yticks(np.arange(0, y_max + 1, 5))

# Griglia orizzontale
ax.set_axisbelow(True)
ax.grid(axis='y', linestyle='--', alpha=0.6)

# Titoli e label
ax.set_xlabel('Durata partita (minuti)')
ax.set_ylabel('Numero di partite')
ax.set_title('Distribuzione della durata delle partite – LEC 2025')

# Box testuale con supporto quantitativo (>45 min)
info = (
    f"Totale partite: {n_total}\n"
    f"> 45 min: {n_over_45} ({pct_over_45:.1f}%)"
)

ax.text(
    0.98, 0.95, info,
    transform=ax.transAxes,
    ha='right', va='top',
    bbox=dict(boxstyle='round', alpha=0.2)
)

ax.legend()
plt.show()
```

### TESTO UTILE
La durata media delle partite LEC 2025 è pari a circa 33 minuti.
Tuttavia, la distribuzione delle durate risulta influenzata dalla presenza di 15 partite particolarmente lunghe (oltre i 45 minuti), che contribuiscono ad aumentare il valore medio.
La maggiore concentrazione di partite si osserva in un intervallo compreso approssimativamente tra 27 e 32 minuti, indicando che la durata tipica di una partita risulta inferiore alla media complessiva.

Combinando queste evidenze con il numero medio di circa 27 kill per partita, emerge un ritmo di gioco pari a circa 0,8 kill al minuto.
Nel complesso, questi risultati suggeriscono che il LEC 2025 presenti caratteristiche compatibili con una lega prevalentemente early–mid game oriented, con un ritmo di gioco sostenuto e una frequente interazione tra le squadre già nelle fasi iniziali e intermedie della partita.

**Qual è la percentuale di vittorie BLUE SIDE vs RED SIDE nel LEC 2025?**

```python
df_side = lec_df.loc[lec_df['position'] == 'team', ['gameid', 'split', 'side', 'result']].copy()
df_side.head(10)
```

```python
df_side.groupby('gameid')['side'].nunique().value_counts()
```

```python
summary_side_split = (
    df_side
    .groupby(['split', 'side'])
    .agg(
        partite=('gameid', 'nunique'),
        vittorie=('result', 'sum')
    )
    .reset_index()
)

dfpr = (
    df_side
    .groupby(['split', 'side'])
)

summary_side_split['win_rate'] = (
    summary_side_split['vittorie'] /
    summary_side_split['partite']
).round(2)
```

```python
summary_side_split.head(10)
```

```python
plt.figure(figsize=(8, 5))
ax = sns.barplot(
    data=summary_side_split,
    x='split',
    y='win_rate',
    hue='side'
)

ax.set_ylabel('Win rate')
ax.set_title('Win rate Blue vs Red per Split – LEC 2025')
ax.set_ylim(0, 1)

# ✅ aggiunta etichette percentuali
for container in ax.containers:
    ax.bar_label(
        container,
        labels=[f'{v*100:.1f}%' for v in container.datavalues],
        padding=3,
        fontsize=10
    )

plt.show()
```

### TESTO UTILE
Per analizzare il win rate per side sono state considerate esclusivamente le righe team-level, in modo da includere entrambi i side per ogni partita.
Dallo Spring split emerge un vantaggio del Blue side, con una percentuale di vittorie superiore di circa 20 punti percentuali rispetto al Red side.
Nel Winter split le differenze risultano più contenute, suggerendo un maggiore equilibrio tra i due side.
Nel Summer split si osserva una leggera inversione, con un win rate del Red side marginalmente superiore a quello del Blue side, sebbene la differenza risulti limitata.

```python
df_winsidelength = (
    lec_df
    .loc[lec_df['position'] == 'team', ['gameid', 'side', 'gamelength', 'result']]
    .copy()
)

df_winsidelength['gamelength_min'] = df_winsidelength['gamelength'] / 60
df_winsidelength['duration_bin'] = pd.qcut(
    df_winsidelength['gamelength_min'],
    q=6,
    precision = 2
)
```

```python
winrate_by_len_side = (
    df_winsidelength
    .dropna(subset=['duration_bin'])
    .groupby(['duration_bin', 'side'], observed = True)
    .agg(
        partite=('gameid', 'nunique'),
        win_rate=('result', 'mean')
    )
    .reset_index()
)

winrate_by_len_side.head(12)
```

```python
plt.figure(figsize=(10, 5))
ax = sns.barplot(
    data=winrate_by_len_side,
    x='duration_bin',
    y='win_rate',
    hue='side'
)
ax.set_ylim(0, 1)
plt.xticks(rotation=45)
plt.tight_layout()

for container in ax.containers:
    ax.bar_label(
        container,
        labels=[f'{v*100:.1f}%' for v in container.datavalues],
        padding=3,
        fontsize=10
    )

    ax.set_title('Win rate Blue vs Red 2025 per Fascia di durata')
    
plt.show()
```

### TESTO UTILE
Per valutare se la durata delle partite influisca sul win rate dei side, sono state considerate esclusivamente le righe team-level, in modo da avere due osservazioni per ogni partita, corrispondenti a Blue side e Red side.
La durata delle partite è stata convertita in minuti e successivamente suddivisa in sei fasce basate sui quantili della distribuzione, così da ottenere gruppi con un numero simile di partite.
Per ciascuna fascia è stato quindi calcolato il win rate dei due side.

Dal grafico emerge che in 5 fasce su 6 il Blue side presenta un win rate superiore, generalmente con un margine ben definito.
Fa eccezione la fascia 28.55–30.56 minuti, in cui il vantaggio del Blue side risulta più contenuto (circa 4 punti percentuali).
Nella fascia di durata più elevata, invece, si osserva un’inversione di tendenza, con il Red side in vantaggio di circa 6 punti percentuali.

**Qual è la frequenza di partite "snowball" (vantaggio gold > X a 15min)?**

```python
snowball_df = (
    lec_df
    .loc[lec_df['position'] == 'team',
         ['gameid', 'golddiffat15']]
    .drop_duplicates('gameid')
    .copy()
)
```

```python
sb_treshold = 3000
snowball_df['snowball'] = (
    snowball_df['golddiffat15'].abs() > sb_treshold
)
labels = ['Snowball', 'Non snowball']
counts = [
    snowball_df['snowball'].sum(),
    (~snowball_df['snowball']).sum()
]

total = sum(counts)
percentages = [c / total * 100 for c in counts]

colors = ['#1f77b4', '#ff7f0e']  # puoi cambiarli se vuoi

# Donut chart
fig, ax = plt.subplots(figsize=(6, 6))

wedges, _ = ax.pie(
    counts,
    labels=None,              # ❌ niente label sulle fette
    colors=colors,
    startangle=90,
    wedgeprops=dict(width=0.4, edgecolor='white')
)

# Legenda custom con percentuali
legend_labels = [
    f'{labels[i]} — {percentages[i]:.1f}%'
    for i in range(len(labels))
]

ax.legend(
    wedges,
    legend_labels,
    title='Tipo di partita',
    loc='center',
    bbox_to_anchor=(0.5, 0.5),   # centro del donut
    frameon=False
)

ax.set_title('Frequenza di partite snowball – LEC 2025')

plt.show()
```

### TESTO UTILE
Considerando una definizione di snowball pari a un vantaggio in gold superiore a 3000 unità entro i primi 15 minuti, risulta che circa il 15% delle partite del LEC 2025 rientra in questa categoria.
Questo suggerisce che, nella maggior parte dei match, il gioco rimane relativamente equilibrato nelle fasi iniziali, con un numero limitato di partite caratterizzate da un forte vantaggio economico da parte di uno dei due team.

**Quante partite vengono chiuse senza Baron o senza Dragon Soul?**

```python
objectives_df = (
    lec_df
    .loc[lec_df['position'] == 'team',
         ['gameid', 'dragons', 'firstbaron']]
    .copy()
)
objectives_df['dragonsoul'] = objectives_df['dragons'] >= 4
objectives_df['major_objective'] = (
    objectives_df['firstbaron'] | objectives_df['dragonsoul']
)
objectives_df.head()
```

```python
game_objectives = (
    objectives_df
    .groupby('gameid')['major_objective']
    .any()
    .reset_index(name='any_major_objective')
)
game_objectives['no_major_objectives'] = (
    ~game_objectives['any_major_objective']
)
```

```python
# -------------------------
# DATI
# -------------------------
labels = ['Senza Baron o Dragon Soul', 'Con Baron o Dragon Soul']
counts = [
    game_objectives['no_major_objectives'].sum(),
    (~game_objectives['no_major_objectives']).sum()
]

total = sum(counts)
pct1 = counts[0] / total * 100
pct2 = counts[1] / total * 100

# -------------------------
# COLORI
# -------------------------
violet = '#7b5fc3'
grey = '#bdbdbd'
colors = [violet, grey]

# -------------------------
# DONUT
# -------------------------
fig, ax = plt.subplots(figsize=(6, 6))

ax.pie(
    counts,
    labels=None,
    colors=colors,
    startangle=90,
    wedgeprops=dict(width=0.4, edgecolor='white')
)

ax.set_title('Partite concluse senza Baron o Dragon Soul – LEC 2025', pad=20)

# -------------------------
# "LEGENDA" CUSTOM CENTRATA
# -------------------------
# Coordinate in ax.transAxes: (0,0)=basso-sx, (1,1)=alto-dx
x0, y0 = 0.5, 0.5

# Layout
line_h = 0.065        # distanza verticale tra righe
box_w, box_h = 0.04, 0.025
x_shift = 0.02        # separazione orizzontale tra % e rettangolo

# Titolo legenda
ax.text(
    x0, y0 + 2.1*line_h,
    "Tipo di partita",
    ha='center', va='center',
    fontsize=13, fontweight='bold',
    transform=ax.transAxes
)

# --- BLOCCO 1 (Snowball/No Baron+Soul) ---
# percentuale (a sinistra del centro)
ax.text(
    x0 - x_shift, y0 + 0.8*line_h,
    f"{pct1:.1f}%",
    ha='right', va='center',
    fontsize=12,
    transform=ax.transAxes
)

# rettangolo (a destra del centro)
ax.add_patch(mpatches.Rectangle(
    (x0 + x_shift, y0 + 0.8*line_h - box_h/2),
    box_w, box_h,
    transform=ax.transAxes,
    facecolor=violet,
    edgecolor='white'
))

# descrizione (riga sotto)
ax.text(
    x0, y0 + 0.05*line_h,
    labels[0],
    ha='center', va='center',
    fontsize=12,
    transform=ax.transAxes
)

# --- BLOCCO 2 (Con Baron o Dragon Soul) ---
ax.text(
    x0 - x_shift, y0 - 1.05*line_h,
    f"{pct2:.1f}%",
    ha='right', va='center',
    fontsize=12,
    transform=ax.transAxes
)

ax.add_patch(mpatches.Rectangle(
    (x0 + x_shift, y0 - 1.05*line_h - box_h/2),
    box_w, box_h,
    transform=ax.transAxes,
    facecolor=grey,
    edgecolor='white'
))

ax.text(
    x0, y0 - 1.8*line_h,
    labels[1],
    ha='center', va='center',
    fontsize=12,
    transform=ax.transAxes
)

plt.show()
```

### TESTO UTILE
Il 92% delle partite del LEC 2025 si conclude con l’ottenimento di almeno uno tra first Baron e Dragon Soul, evidenziando l’importanza degli obiettivi maggiori nella risoluzione dei match.

In altre parole, la competizione sembra essersi strutturata attorno a pochi team in grado di imporre sistematicamente un vantaggio nelle fasi iniziali della partita, piuttosto che su un equilibrio diffuso tra tutte le squadre.

**Esiste una relazione tra gold diff @15 e win rate finale del team?**

```python
team_gold_win = lec_df[
    ['position', 'teamname', 'golddiffat15', 'result']
]

team_gold_win = team_gold_win[
    team_gold_win['position'] == 'team'
]

team_gold_win = (
    team_gold_win
    .groupby(['teamname', 'result'], as_index=False)
    .agg(
        avg_golddiffat15=('golddiffat15', 'mean'),
        games=('golddiffat15', 'count')
    )
)

team_gold_win_pivot = team_gold_win.pivot(
    index='teamname',
    columns='result',
    values='avg_golddiffat15'
).reset_index()

team_gold_win_pivot.columns.name = None
team_gold_win_pivot = team_gold_win_pivot.rename(columns={
    False: 'avg_gold_loss',
    True: 'avg_gold_win'
})
team_gold_win_pivot.head(10)
```

```python
df = team_gold_win_pivot.sort_values('avg_gold_win')

plt.figure(figsize=(10, 6))

plt.hlines(
    y=df['teamname'],
    xmin=df['avg_gold_loss'],
    xmax=df['avg_gold_win']
)

plt.scatter(df['avg_gold_loss'], df['teamname'])
plt.scatter(df['avg_gold_win'], df['teamname'])

plt.axvline(0)

plt.xlabel('Gold Difference medio a 15 minuti')
plt.title('Gold diff @15: confronto Vittorie vs Sconfitte per Team – LEC 2025')

plt.tight_layout()
plt.show()
```

l grafico mostra la relazione tra il gold difference medio a 15 minuti e l’esito finale della partita per ciascun team, distinguendo tra vittorie e sconfitte.

Nel complesso, si osserva che tutti i team presentano un vantaggio economico early game nettamente positivo nelle partite vinte, mentre mostrano valori negativi o prossimi allo zero nelle sconfitte, evidenziando una forte associazione tra early advantage e outcome finale.

Emergono tuttavia due casi particolari. Natus Vincere non presenta osservazioni di vittoria; di conseguenza, il confronto è limitato alle sole sconfitte, che risultano associate a un marcato svantaggio economico già nei primi 15 minuti.

Al contrario, G2 Esports rappresenta un’eccezione rilevante: è l’unico team che, anche nelle partite perse, mantiene in media un gold difference positivo a 15 minuti, suggerendo una capacità di generare vantaggio early game indipendentemente dall’esito finale della partita.

**Quali team convertono meglio il vantaggio iniziale in vittoria?**

```python
conversion_df = lec_df[
    ['position', 'teamname', 'golddiffat15', 'result']
]

conversion_df = conversion_df[
    conversion_df['position'] == 'team'
]
conversion_df['early_advantage'] = conversion_df['golddiffat15'] > 0
conversion_df = conversion_df[
    conversion_df['early_advantage']
]
team_conversion = (
    conversion_df
    .groupby('teamname')
    .agg(
        games_with_adv=('result', 'count'),
        wins_with_adv=('result', 'sum')
    )
    .reset_index()
)

team_conversion['conversion_rate'] = (
    team_conversion['wins_with_adv'] /
    team_conversion['games_with_adv']
)

# Ordiniamo i team per conversion rate
plot_df = team_conversion.sort_values(
    'conversion_rate',
    ascending=False
)

plt.figure(figsize=(10, 6))

plt.barh(
    plot_df['teamname'],
    plot_df['conversion_rate']
)

plt.xlabel('Conversion rate (win | gold diff @15 > 0)')
plt.title('Capacità di conversione del vantaggio early in vittoria – LEC 2025')

# Linea di riferimento al 50%
plt.axvline(0.5, linestyle='--')

# Annotazioni con numero di partite
for i, row in plot_df.iterrows():
    plt.text(
        row['conversion_rate'] + 0.01,
        row['teamname'],
        f"{int(row['games_with_adv'])}",
        va='center'
    )

plt.xlim(0, 1)
plt.tight_layout()
plt.show()
```

### TESTO UTILE
Dal grafico emerge che Fnatic, G2 Esports e Team Vitality sono i team che mostrano la maggiore capacità di conversione del vantaggio economico iniziale in vittoria.

Tra questi, Fnatic si distingue come il team più efficiente, riuscendo a convertire oltre l’80% delle partite in cui si trova in vantaggio a 15 minuti in una vittoria finale.

G2 Esports e Team Vitality presentano anch’essi un’elevata capacità di conversione, sebbene con percentuali leggermente inferiori rispetto a Fnatic.

**Quali team recuperano più spesso partite partendo in svantaggio?**

```python
recovery_df = lec_df[
    ['position', 'teamname', 'golddiffat15', 'result']
]

recovery_df = recovery_df[
    recovery_df['position'] == 'team'
]

recovery_df['early_disadvantage'] = recovery_df['golddiffat15'] < 0

recovery_df = recovery_df[
    recovery_df['early_disadvantage']
]

team_recovery = (
    recovery_df
    .groupby('teamname')
    .agg(
        games_from_behind=('result', 'count'),
        wins_from_behind=('result', 'sum')
    )
    .reset_index()
)

team_recovery['recovery_rate'] = (
    team_recovery['wins_from_behind'] /
    team_recovery['games_from_behind']
)

plot_df = team_recovery.sort_values(
    'recovery_rate',
    ascending=False
)

plt.figure(figsize=(10, 6))

plt.barh(
    plot_df['teamname'],
    plot_df['recovery_rate']
)

plt.xlabel('Recovery rate (win | gold diff @15 < 0)')
plt.title('Capacità di recupero partendo in svantaggio – LEC 2025')

# Linea di riferimento al 50%
plt.axvline(0.5, linestyle='--')

# Annotazioni con numero di partite giocate in svantaggio
for i, row in plot_df.iterrows():
    plt.text(
        row['recovery_rate'] + 0.01,
        row['teamname'],
        f"{int(row['games_from_behind'])}",
        va='center'
    )

plt.xlim(0, 1)
plt.tight_layout()
plt.show()
```

### TESTO UTILE
Dal grafico emerge che la capacità di recuperare una partita partendo in svantaggio a 15 minuti è generalmente bassa per la maggior parte dei team.

La quasi totalità delle squadre presenta infatti un recovery rate inferiore al 50%, indicando che, nella maggior parte dei casi, uno svantaggio early game tende a tradursi in una sconfitta finale.

Fanno eccezione Karmine Corp e G2 Esports, che risultano essere gli unici team in grado di ribaltare lo svantaggio iniziale in oltre il 50% delle occasioni, evidenziando una maggiore efficacia nelle fasi di mid e late game.
