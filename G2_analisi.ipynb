### ANALISI GENERALI - CONTESTO COMPETITIVO
**Quante partite sono state giocate nel LEC nel 2025?**

```python
df['league'].unique()
```

```python
lec_df = df[df['league'] == 'LEC']
lec_df['split'].unique()
```

Nel dataset LEC 2025 sono presenti i seguenti split: Winter, Spring, Summer

```python
games_per_split = (
    lec_df
    .drop_duplicates(subset='gameid')
    .groupby('split')['gameid']
    .count()
)
total_games = games_per_split.sum()

print(games_per_split)
print(total_games)
```

Durante il 2025 sono state giocate:
* 139 partite in Spring Split
* 83 partite in Summer Split
* 84 partite in Winter Split

### TESTO UTILIE
Poiché ogni partita è identificata univocamente da gameid ma compare più volte nel dataset per rappresentare statistiche di team e giocatori, il numero di partite è stato calcolato considerando i valori unici di gameid.
Successivamente, ogni partita è stata assegnata al relativo split (Spring, Summer, Finals) e aggregata per ottenere la distribuzione delle partite nel corso della stagione 2025.


```python
summary_gamesPerSplit_df = games_per_split.reset_index()
summary_gamesPerSplit_df.columns = ['split', 'games']
total_games = summary_gamesPerSplit_df['games'].sum()

summary_gamesPerSplit_df['percentage (%)'] = (
    summary_gamesPerSplit_df['games'] / total_games * 100
).round(2)

summary_gamesPerSplit_df.head()
```

```python

labels = summary_gamesPerSplit_df['split']
sizes = summary_gamesPerSplit_df['games']

fig, ax = plt.subplots(figsize=(7, 7))

ax.pie(
    sizes,
    labels=labels,
    autopct='%1.1f%%',
    startangle=90,
    wedgeprops={'width': 0.4},
    pctdistance=0.80
)

# Numero totale al centro
ax.text(
    0, 0,
    f'{total_games}\nTotal Games',
    ha='center',
    va='center',
    fontsize=12,
    fontweight='bold'
)

ax.set_title(
    'Distribuzione delle partite LEC 2025 per Split',
    fontsize=14
)

plt.show()
```

**Qual è stata la durata media delle partite LEC nel 2025?**

```python
games_duration_df = lec_df.drop_duplicates(subset='gameid').copy()

games_duration_df['gamelength_min'] = games_duration_df['gamelength'] / 60
mean_duration = games_duration_df['gamelength_min'].mean()
median_duration = games_duration_df['gamelength_min'].median()

n_total = games_duration_df['gamelength_min'].shape[0]
n_over_45 = (games_duration_df['gamelength_min'] > 45).sum()
pct_over_45 = n_over_45 / n_total * 100
```

```python
plt.figure(figsize=(8, 5))
ax = sns.histplot(games_duration_df['gamelength_min'], bins=20, kde=True)

# Linee verticali: media e mediana
ax.axvline(mean_duration, linestyle='--', linewidth=2, label=f"Media: {mean_duration:.1f} min")
ax.axvline(median_duration, linestyle=':', linewidth=2, label=f"Mediana: {median_duration:.1f} min")

# Asse Y con step fisso (come hai già fatto)
y_max = ax.get_ylim()[1]
ax.set_yticks(np.arange(0, y_max + 1, 5))

# Griglia orizzontale
ax.set_axisbelow(True)
ax.grid(axis='y', linestyle='--', alpha=0.6)

# Titoli e label
ax.set_xlabel('Durata partita (minuti)')
ax.set_ylabel('Numero di partite')
ax.set_title('Distribuzione della durata delle partite – LEC 2025')

# Box testuale con supporto quantitativo (>45 min)
info = (
    f"Totale partite: {n_total}\n"
    f"> 45 min: {n_over_45} ({pct_over_45:.1f}%)"
)

ax.text(
    0.98, 0.95, info,
    transform=ax.transAxes,
    ha='right', va='top',
    bbox=dict(boxstyle='round', alpha=0.2)
)

ax.legend()
plt.show()
```

### TESTO UTILE
La durata media delle partite LEC 2025 è pari a circa 33 minuti.
Tuttavia, la distribuzione delle durate risulta influenzata dalla presenza di 15 partite particolarmente lunghe (oltre i 45 minuti), che contribuiscono ad aumentare il valore medio.
La maggiore concentrazione di partite si osserva in un intervallo compreso approssimativamente tra 27 e 32 minuti, indicando che la durata tipica di una partita risulta inferiore alla media complessiva.

**La durata media delle partite LEC  è variata nei vari split?**

```python
summary_split_df = (
    games_duration_df
    .groupby('split')['gamelength_min']
    .agg(
        media='mean',
        mediana='median',
        std='std',
        min='min',
        max='max',
        partite='count'
    )
    .round(2)
)
```

```python
summary_split_df.head()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(
    data=games_duration_df,
    x='split',
    y='gamelength_min'
)

plt.xlabel('Split')
plt.ylabel('Durata partita (minuti)')
plt.title('Distribuzione della durata delle partite per Split – LEC 2025')
plt.show()
```

### TESTO UTILE
Dai boxplot emerge che la mediana della durata delle partite risulta simile nei tre split, indicando che la durata tipica delle partite non varia in modo sostanziale tra Winter, Spring e Summer.
Tuttavia, lo split Winter presenta una maggiore variabilità, evidenziata da un intervallo interquartile più ampio e dalla presenza di valori estremi più distanti rispetto agli altri due split.
Sebbene le mediane risultino simili, le durate medie mostrano lievi differenze tra gli split, suggerendo che la variabilità osservata – in particolare nello split Winter – contribuisca a influenzare il valore medio.

**Qual è la percentuale di vittorie BLUE SIDE vs RED SIDE nel LEC 2025?**

```python
df_side = lec_df.loc[lec_df['position'] == 'team', ['gameid', 'split', 'side', 'result']].copy()
df_side.head(10)
```

```python
df_side.groupby('gameid')['side'].nunique().value_counts()
```

```python
summary_side_split = (
    df_side
    .groupby(['split', 'side'])
    .agg(
        partite=('gameid', 'nunique'),
        vittorie=('result', 'sum')
    )
    .reset_index()
)

dfpr = (
    df_side
    .groupby(['split', 'side'])
)

summary_side_split['win_rate'] = (
    summary_side_split['vittorie'] /
    summary_side_split['partite']
).round(2)
```

```python
summary_side_split.head(10)
```

```python
plt.figure(figsize=(8, 5))
ax = sns.barplot(
    data=summary_side_split,
    x='split',
    y='win_rate',
    hue='side'
)

ax.set_ylabel('Win rate')
ax.set_title('Win rate Blue vs Red per Split – LEC 2025')
ax.set_ylim(0, 1)

# ✅ aggiunta etichette percentuali
for container in ax.containers:
    ax.bar_label(
        container,
        labels=[f'{v*100:.1f}%' for v in container.datavalues],
        padding=3,
        fontsize=10
    )

plt.show()
```

### TESTO UTILE
Per analizzare il win rate per side sono state considerate esclusivamente le righe team-level, in modo da includere entrambi i side per ogni partita.
Dallo Spring split emerge un vantaggio del Blue side, con una percentuale di vittorie superiore di circa 20 punti percentuali rispetto al Red side.
Nel Winter split le differenze risultano più contenute, suggerendo un maggiore equilibrio tra i due side.
Nel Summer split si osserva una leggera inversione, con un win rate del Red side marginalmente superiore a quello del Blue side, sebbene la differenza risulti limitata.
