# Progetto G2 - Documento di pianificazione

## 1. Contesto del progetto
Questo progetto ha l’obiettivo di analizzare un database contenente tutte le partite di **League of Legends giocate nel 2025 dalla squadra G2 Esport durante l'anno competitivo 2025**. L’analisi è pensata come **progetto personale da portfolio** per un ruolo di Data Analyst e **segue il framework PACE (Plan, Analyze, Construct, Evaluate)**.

Il progetto simula un contesto reale di analisi sportiva/esport, dove i dati vengono utilizzati per valutare performance, individuare punti di forza e debolezza del team e supportare decisioni strategiche.

### Fonte dei dati
Per motivi di **licenza** e di utilizzo delle piattaforme di distribuzione dei dati, **il file CSV originale non viene allegato al progetto**, i dati tuttavia sono pubblicamente reperibili e scaricati gratuitamente dalla seguente fonte:

https://oracleselixir.com/

## 2. Audience di riferimento
Questo progetto è realizzato **a scopo pratico e formativo** ed è destinato all’inserimento in un **portfolio personale da Data Analyst**. Non esiste quindi una committenza reale né un’audience operativa diretta.

Tuttavia, per simulare un contesto professionale realistico e dimostrare la capacità di adattare l’analisi a diversi stakeholder, viene definita un’audience teorica di riferimento.

**Audience primaria (teorica):**
* Recruiter e Hiring Manager per ruoli di **Data Analyst**
* Data Analytics Manager

**Audience secondaria (teorica):**
* Coaching staff della squadra
* Team manager
* Analisti esport

Il livello di comunicazione è pensato per essere **data-driven ma accessibile**, con visualizzazioni chiare e insight spiegati in modo comprensibile anche a un pubblico non tecnico, replicando un tipico scenario aziendale.

## 3. Prima fase - PLAN
La prima tappa del progetto all’interno del framework **PACE** è la fase di **Plan (Pianificazione)**. In questa fase vengono definite le fondamenta dell’intero lavoro analitico: le risorse necessarie, gli obiettivi, le domande chiave e l’impatto atteso.

Una pianificazione chiara consente di orientare correttamente l’analisi, evitare esplorazioni non mirate e garantire che le successive fasi del progetto producano insight coerenti e rilevanti rispetto agli obiettivi iniziali.

### 3.1 Risorse necessarie
Per lo svolgimento della fase di pianificazione e delle successive fasi del progetto sono necessarie le seguenti risorse tecniche e informative:

**Dati**
* Database contenente le partite di League of Legends disputate dalla squadra durante l’anno 2025, con informazioni generali sul match e specifiche sulle performance dei singoli giocatori.

**Strumenti di analisi:**
* **Python 3**, linguaggio di programmazione che verrà utilizzato durante tutto lo sviluppo del progetto
* **Numpy**, framework Python utilizzato per le operazione matematiche sui dati
* **Pandas**, framework Python utilizzato per gestione e manipolazione del dataset
* **Seaborn**, framework Python utilizzato per creare visualizzazioni grafiche esplorative e comparative (il framework potrebbe essere sostituito dal software **Tableau** per le visualizzazioni inserite nel rapporto finale)

Queste risorse costituiscono la base tecnica su cui verranno costruite tutte le analisi, le visualizzazioni e gli eventuali modelli sviluppati nelle fasi successive del framework PACE.

### 3.2 Obiettivi del progetto
Questa sezione definisce in modo esplicito cosa si intende ottenere dall’analisi dei dati. Gli obiettivi sono formulati per essere misurabili, verificabili e direttamente collegati alle metriche disponibili nel dataset, così da guidare in modo coerente le attività delle fasi successive del framework PACE.

**Obiettivi analitici principali:**
* Analizzare le **performance medie dei singoli giocatori** del team sull'intero periodo considerato
* Confrontare le **performance dei giocatori in caso di vittoria e in caso di sconfitta** per scoprire se ci sono dei pattern su specifici intervalli della partita da approfondire per eventuali miglioramenti
* Individuare quali giocatori e/o ruoli risultano **maggiormente associati alle vittorie**
* Individuare quali giocatori e/o ruoli risultano **maggiormente associati alle sconfitte**
* Valutare l’**impatto degli obiettivi neutrali** (draghi, Herald, Baron) sull’esito delle partite

Nel corso dell'analisi gli obiettivi potranno essere progressivamente **adattati o approfanditi** durante la fase di analisi, qualora emergano nuovi pattern o domande rilevanti dai dati.

### 3.3 Domande chiave, quesiti analitici e ipotesi
Questa sezione raccoglie in modo unificato le domande chiave, i **quesiti di business/analitici** e le **ipotesi iniziali** che guideranno l’intera fase di analisi. L’obiettivo è fornire una visione chiara e coerente delle domande poste ai dati e delle assunzioni che verranno successivamente verificate.

**domande chiave e quesiti analitici**
1. Quali sono le **performance medie** dei giocatori del team nel periodo analizzato? E come si relano se confrontate con quelle di altri team?
2. In che modo le **performance individuali** cambiano tra **vittorie e sconfitte**?
3. Qual è il **win rate complessivo** della squadra nel 2025?
4. Il team performa meglio in **early, mid o late game**?
5. Esistono **pattern ricorrenti nelle sconfitte**?
6. Esistono metriche che risultano **fortemente associate all’esito della partita**?
7. Alcuni ruoli (top, jungle, mid, adc, support) hanno un **impatto maggiore** sull’esito?
8. In che misura il **controllo degli obiettivi neutrali** influenza il risultato finale?
9. Il risultato varia in base al **side** (blue/red)?

**Ipotesi iniziali**
* Le **vittorie** sono associate a un **vantaggio economico già nei primi 15 minuti**
* Le **sconfitte** presentano una perdita **sistematica di controllo degli obiettivi neutrali**
* Il ruolo del **jungler ha un impatto particolarmente rilevante** nelle partite vinte
* Alcuni giocatori o campioni mostrano **performance superiori alla media** in modo ricorrente

Queste ipotesi verranno esplorate e, ove possibile, verificate empiricamente durante la fase **Analyze**.

### 3.4 Impatto atteso
Se questo progetto fosse realizzato per un contesto reale competitivo o professionale, l’impatto includerebbe:
* Supporto alle decisioni strategiche del team
* Migliorare comprensione delle condizioni che portano alla vittoria
* Identificazione delle aree critiche su cui focalizzare allenamento e preparazione

Nel contesto del portfolio personale, l’impatto del progetto è:
* Dimostrare competenze di **analisi dati end-to-end**
* Evidenziare capacità di **pensiero critico e analitico**
* Mostrare l’abilità di tradurre dati complessi in **insight chiari e utilizzabili**

## 4. Seconda fase - ANALYZE
In questa fase il data analyst entra in contatto diretto con i dati, un passaggio cruciale per trasformare le informazioni grezze in insight significativi.

L’analisi dei dati consente di comprendere la struttura del dataset, valutarne la qualità, individuare pattern rilevanti e verificare le ipotesi definite nella fase di pianificazione, ponendo le basi per le successive attività di costruzione e comunicazione dei risultati.

### 4.1 Interazione con i dati
In questa sottofase avviene l’acquisizione dei dati all’interno dell’ambiente di lavoro. I dati vengono importati, caricati e resi disponibili per le successive attività di esplorazione e analisi.

Questa fase rappresenta il primo contatto operativo con il dataset si selezionano i dati utili all'analisi e garantisce che le informazioni siano correttamente accessibili, leggibili e pronte per essere analizzate in modo sistematico.

### 4.2 EDA - Exploratory data analysis
In questa fase i dati vengono esplorati in modo sistematico per comprenderne struttura, qualità e distribuzioni principali, l’obiettivo non è trarre conclusioni definitive, ma acquisire familiarità con il dataset e individuare eventuali criticità o pattern preliminari.

**Attività tipiche dell’EDA:**
* **Formattazione del database:** organizzazione dei dati in una struttura coerente e facilmente analizzabile.
* **Conversione dei dati in formati utilizzabili:** trasformazione dei dati grezzi in formati idonei all’analisi.
* **Pulizia dei dati:** individuazione e gestione di errori, valori incoerenti o dati non pertinenti.
* **Analisi della struttura del dataset:** verifica del numero di righe e colonne e comprensione della granularità dei dati.
* **Ispezione dei tipi di dato e delle variabili disponibili:** controllo della corretta assegnazione dei tipi di dato.
* **Identificazione di valori mancanti o incompleti:** valutazione della presenza di dati nulli o parziali.
* **Analisi delle distribuzioni delle variabili numeriche:** studio delle distribuzioni per comprendere comportamenti tipici e anomalie.
* **Individuazione di outlier o valori anomali:** rilevazione di osservazioni estreme.
* **Prime visualizzazioni esplorative:** utilizzo di grafici preliminari per individuare trend generali e pattern iniziali.

### 4.3 Fase di analisi
In questa fase **i dati grezzi vengono analizzati in modo strutturato** attraverso il calcolo di metriche e statistiche descrittive, con l’obiettivo di sintetizzare e organizzare le informazioni contenute nel dataset. **I risultati ottenuti vengono quindi interpretati** e trasformati in insight utili a supporto delle domande analitiche, semplificandone la comprensione **mediante l’uso di visualizzazioni grafiche**.

L’analisi si focalizza in particolare sul confronto tra partite vinte e perse, sulle performance individuali e di ruolo e sul controllo degli obiettivi, fornendo una base solida per l’interpretazione finale dei risultati.

### 4.4 Sintesi e insight
I risultati dell’analisi vengono **sintetizzati** in forma strutturata, **evidenziando pattern ricorrenti**, punti di forza e debolezza del team e possibili aree di approfondimento per le fasi successive.

La fase Analyze fornisce le basi quantitative e qualitative per la costruzione delle visualizzazioni e delle analisi avanzate previste nelle fasi successive del framework PACE

## 5. Terza fase - CONSTRUCT
Questa fase è dedicata alla **creazione di modelli descrittivi o predittivi** basati su tecniche di machine learning che consentono di **analizzare il legame tra variabili di input** (performance individuali, controllo degli obiettivi, vantaggi temporali) **e risultati di output**, come l’esito delle partite, supportando una **comprensione più profonda** dei fattori che influenzano le performance del team.

### 5.1 Selezione e preparazione delle feature
In questa sottofase vengono **selezionati, trasformati e preparati i dati di input** (feature)che alimenteranno i modelli descrittivi o predittivi. 

### 5.2 Costruzione dei modelli
In questa sottofase vengono costruiti modelli descrittivi o predittivi utilizzando tecniche di machine learning, con l’obiettivo di individuare relazioni complesse tra le variabili di input e l’esito delle partite.

### 5.3 Interpretazione dei modelli

I risultati prodotti dai modelli vengono analizzati e interpretati per comprendere quali fattori contribuiscono maggiormente all’esito delle partite. Questa sottofase consente di tradurre l’output dei modelli in insight comprensibili e utilizzabili.

## 6. Quarta fase - EXECUTE
In questa fase vengono **comunicati e messi in pratica i risultati delle analisi e dei modelli** sviluppati nelle fasi precedenti. L’obiettivo è trasformare gli insight analitici in valore concreto, assicurando che i risultati siano compresi, discussi e utilizzabili dai diversi stakeholder.

### 6.1 Presentazione dei risultati
In questa sottofase i risultati finali dell’analisi e dei modelli vengono presentati in modo chiaro e strutturato. La comunicazione è adattata all’audience di riferimento e supportata da visualizzazioni, sintesi testuali e indicatori chiave.

**Le attività includono:**

* Presentazione degli insight principali

* Spiegazione dei risultati dei modelli in modo comprensibile

* Evidenziazione delle implicazioni pratiche per il team

### 6.2 Collezione del feedback

In questa sottofase vengono raccolti i **feedback degli stakeholder**, con l’obiettivo di valutare la chiarezza, la rilevanza e l’utilità dei risultati presentati.

**Il feedback può riguardare:**

* Interpretazione dei risultati

* Coerenza degli insight con il contesto di gioco

* Necessità di ulteriori analisi o approfondimenti

### 6.3 Iterazioni basate sui feedback
Sulla base dei feedback raccolti, il progetto entra in una fase iterativa in cui le analisi, i modelli o le visualizzazioni possono essere **rivisti e migliorati**. Questo processo consente di affinare i risultati e di aumentare il valore complessivo del lavoro svolto.

**L’iterazione può includere:**

* Aggiornamento delle analisi

* Miglioramento dei modelli

* Raffinamento della comunicazione dei risultati
